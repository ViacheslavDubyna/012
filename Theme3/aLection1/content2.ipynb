### Лекція 3.1. Навчальне питання 2: Методи попереднього аналізу даних.

**Текстовий матеріал:**

**Попередній аналіз даних (Exploratory Data Analysis, EDA)** — це початковий етап роботи з даними, спрямований на їх дослідження, виявлення закономірностей, аномалій та потенційно корисних ознак. Він допомагає краще зрозуміти структуру та особливості даних, що є важливим для подальшого вибору методів обробки та аналізу.

**Цілі попереднього аналізу даних в ІАЗ ОВУ:**

*   **Ознайомлення з даними:** Отримання загального уявлення про дані, їх типи, обсяги, розподіли значень тощо.
*   **Виявлення аномалій та помилок:** Пошук незвичайних значень, викидів, пропусків та інших проблем у даних, які можуть негативно вплинути на аналіз.
*   **Виявлення закономірностей та залежностей:** Пошук кореляцій, трендів та інших взаємозв'язків між ознаками, які можуть бути корисними для аналізу та моделювання.
*   **Формування гіпотез та ідей для подальшого аналізу:** На основі результатів EDA можна сформулювати гіпотези про дані та визначити напрямки подальшого дослідження.

**Методи попереднього аналізу даних:**

*   **Статистичний аналіз:** Обчислення основних статистичних характеристик даних (середнє, медіана, мода, дисперсія, стандартне відхилення тощо).
*   **Візуалізація даних:** Створення графіків, діаграм та інших візуальних представлень даних для їх кращого розуміння та виявлення закономірностей.
*   **Кластеризація:** Групування схожих об'єктів у кластери для виявлення прихованих структур у даних.
*   **Аналіз викидів:** Виявлення аномальних значень, які значно відрізняються від більшості даних.
*   **Аналіз пропусків:** Виявлення та обробка пропущених значень у даних.

**Вплив результатів попереднього аналізу на вибір методів подальшої обробки та аналізу даних:**

*   **Виявлення аномалій та помилок:** Допомагає визначити необхідність очищення та корекції даних перед подальшим аналізом.
*   **Виявлення закономірностей та залежностей:** Допомагає вибрати найбільш підходящі методи моделювання та аналізу даних.
*   **Розуміння розподілів та типів даних:** Допомагає визначити необхідність нормалізації, стандартизації або інших перетворень даних.
*   **Виявлення пропущених значень:** Допомагає вибрати стратегію обробки пропусків (видалення, заповнення тощо).
