# Тема 3. Заняття 4. Методи підготовки даних до аналізу (моделювання)

## Питання 1: Методи трансформації набору даних

### Мета
Ознайомити слухачів із методами трансформації даних, які є ключовими етапами підготовки даних до моделювання.

### 1. Створення тестового набору даних
```python
import pandas as pd
import numpy as np

# Створимо тестовий набір даних для демонстрації
data = {
    'age': [25, 30, 45, 50, np.nan, 23],
    'salary': [50000, 54000, 65000, np.nan, 52000, 48000],
    'experience_years': [1, 3, 10, 7, 5, np.nan]
}
df = pd.DataFrame(data)

# Виведемо початковий набір даних
print("Початковий набір даних:")
display(df)
```

### 2. Робота з пропущеними даними
Заповнення пропусків середніми значеннями.

```python
# Заповнення пропусків середніми значеннями
df_filled = df.fillna(df.mean())

print("\nНабір даних після заповнення пропусків середніми значеннями:")
display(df_filled)
```

### 3. Нормалізація даних
Нормалізація числових значень для зарплат і досвіду роботи.

```python
# Нормалізація числових значень
df_filled['salary_norm'] = (df_filled['salary'] - df_filled['salary'].mean()) / df_filled['salary'].std()
df_filled['experience_norm'] = (df_filled['experience_years'] - df_filled['experience_years'].mean()) / df_filled['experience_years'].std()

print("\nНабір даних після нормалізації:")
display(df_filled)
```

### 4. Перетворення категоріальних змінних (One-Hot Encoding)
Як приклад, додамо новий стовпець із категоріями 'освіта'.

```python
# Уявімо, що у нас є новий стовпець з категоріями, наприклад, 'освіта'
df_filled['education'] = ['bachelor', 'master', 'phd', 'bachelor', 'phd', 'master']

# Використаємо метод One-Hot Encoding
df_transformed = pd.get_dummies(df_filled, columns=['education'])

print("\nНабір даних після перетворення категоріальних змінних:")
display(df_transformed)
```

### 5. Масштабування даних (Min-Max Scaling)
Масштабування стовпців 'salary' та 'experience_years' за допомогою Min-Max Scaling.

```python
from sklearn.preprocessing import MinMaxScaler

# Створимо екземпляр MinMaxScaler
scaler = MinMaxScaler()

# Масштабуємо стовпці 'salary' та 'experience_years'
df_transformed[['salary_scaled', 'experience_scaled']] = scaler.fit_transform(df_transformed[['salary', 'experience_years']])

print("\nНабір даних після масштабування:")
display(df_transformed)
```

### Висновки

1. Заповнення пропущених значень.
2. Нормалізація числових даних.
3. Перетворення категоріальних змінних за допомогою One-Hot Encoding.
4. Масштабування даних за допомогою Min-Max Scaling.
