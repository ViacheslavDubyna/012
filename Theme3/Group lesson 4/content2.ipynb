Ознайомити слухачів з основними методами декомпозиції набору даних для полегшення моделювання та підвищення ефективності моделей.

### 1. Що таке декомпозиція набору даних?
Декомпозиція — це процес розбиття складного набору даних на простіші підмножини для полегшення подальшого аналізу та моделювання. Основними методами декомпозиції є:
- Поділ на навчальну та тестову вибірки.
- Виділення важливих ознак (факторів).
- Агрегація або групування даних за категоріями.

### 2. Поділ на навчальну та тестову вибірки
Поділ даних на дві (або більше) частини дозволяє моделі навчатися на одній частині даних, а тестуватися на іншій, що допомагає оцінити її точність і уникнути перенавчання.

```python
from sklearn.model_selection import train_test_split

# Приклад: Створимо штучний набір даних
data = {
    'age': [25, 30, 45, 50, 23, 40],
    'salary': [50000, 54000, 65000, 70000, 48000, 60000],
    'experience_years': [1, 3, 10, 7, 5, 8]
}
df = pd.DataFrame(data)

# Поділ на навчальну (80%) та тестову (20%) вибірки
train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)

print("Навчальна вибірка:")
display(train_set)

print("\nТестова вибірка:")
display(test_set)
```

### 3. Виділення важливих ознак (Feature Selection)
Для побудови моделі важливо визначити, які ознаки найбільш впливають на прогнозування. Зазвичай застосовують методи кореляції або інші статистичні підходи.

```python
# Використаємо метод кореляції для оцінки зв'язків між ознаками
correlation_matrix = df.corr()

print("Матриця кореляції між ознаками:")
display(correlation_matrix)
```

### 4. Агрегація та групування даних
Групування даних дозволяє зменшити кількість унікальних записів шляхом їх агрегації за певними категоріями або значеннями. Це може бути корисним для зменшення шуму в даних або для аналізу тенденцій.

```python
# Приклад групування за кількістю років досвіду
grouped_data = df.groupby('experience_years').agg({'salary': ['mean', 'count']})

print("Агрегація даних за кількістю років досвіду:")
display(grouped_data)
```
