```python
from graphviz import Digraph

# Створення об'єкту діаграми
dot = Digraph(comment='Контекстна діаграма A-0: Оперативне управління НГУ')
dot.attr(rankdir='LR')  # Орієнтація зліва направо

# Центральний процес
dot.node('A0', 'Здійснення оперативного управління НГУ', shape='box', style='filled', fillcolor='lightgrey')

# Входи (INPUTS)
inputs = {
    'input1': 'Директиви вищого рівня',
    'input2': 'Оперативна обстановка',
    'input3': 'Запити від підрозділів',
    'input4': 'Дані про ресурси',
    'input5': 'Цілі держполітики'
}
for key, label in inputs.items():
    dot.node(key, label, shape='ellipse', fillcolor='lightblue', style='filled')
    dot.edge(key, 'A0')

# Керування (CONTROLS)
controls = {
    'control1': 'Законодавство України',
    'control2': 'Військові статути',
    'control3': 'Стратегічні плани',
    'control4': 'Замисел командування',
    'control5': 'Правила ROE',
    'control6': 'Стандарти НАТО',
    'control7': 'Ресурсні обмеження'
}
for key, label in controls.items():
    dot.node(key, label, shape='diamond', fillcolor='lightyellow', style='filled')
    dot.edge(key, 'A0')

# Виходи (OUTPUTS)
outputs = {
    'output1': 'Управлінські рішення',
    'output2': 'Дії підрозділів НГУ',
    'output3': 'Звіти командуванню',
    'output4': 'Оперативна картина',
    'output5': 'Взаємодія',
    'output6': 'Розпорядження ресурсами'
}
for key, label in outputs.items():
    dot.node(key, label, shape='ellipse', fillcolor='lightgreen', style='filled')
    dot.edge('A0', key)

# Механізми (MECHANISMS)
mechanisms = {
    'mech1': 'Органи управління НГУ',
    'mech2': 'Пункти управління',
    'mech3': 'Система зв’язку',
    'mech4': 'Розвідка',
    'mech5': 'Логістика',
    'mech6': 'Особовий склад',
    'mech7': 'Нормативна база'
}
for key, label in mechanisms.items():
    dot.node(key, label, shape='component', fillcolor='lightpink', style='filled')
    dot.edge(key, 'A0', style='dashed')

# Налаштування стилю
dot.attr(label='Контекстна діаграма A-0: Оперативне управління НГУ\nДата: 07.04.2025')
dot.attr(fontsize='12')

# Збереження та відображення
dot.render('context_diagram_A0', format='png', cleanup=True)
print("Діаграма збережена як 'context_diagram_A0.png'")
```

```python
import os
import sys

def check_system():
    test_dir = r'd:\Work ub\012\Practice 4.3\practice 5'
    try:
        # Check Python version
        print(f"Python version: {sys.version}")
        
        # Check directory existence
        if not os.path.exists(test_dir):
            print(f"Directory doesn't exist: {test_dir}")
            try:
                os.makedirs(test_dir)
                print("Directory created successfully")
            except Exception as e:
                print(f"Cannot create directory: {e}")
                return False
        else:
            print("Directory exists")
        
        # Check write permissions
        test_file = os.path.join(test_dir, 'test.txt')
        try:
            with open(test_file, 'w') as f:
                f.write('test')
            print("Write test: Success")
            
            # Check read permissions
            with open(test_file, 'r') as f:
                content = f.read()
            print("Read test: Success")
            
            # Clean up
            os.remove(test_file)
            print("Delete test: Success")
            
            return True
        except Exception as e:
            print(f"File operation failed: {e}")
            return False
            
    except Exception as e:
        print(f"System check failed: {e}")
        return False

# Run the check
if check_system():
    print("\nAll system checks passed ✅")
else:
    print("\nSystem check failed ❌")
```

```python
import os
import sys
import subprocess

def перевірка_системи():
    тест_директорія = r'd:\Work ub\012\Practice 4.3\practice 5'
    try:
        # Перевірка версії Python
        print(f"Версія Python: {sys.version}")
        
        # Перевірка наявності директорії
        if not os.path.exists(тест_директорія):
            print(f"Директорія не існує: {тест_директорія}")
            try:
                os.makedirs(тест_директорія)
                print("Директорію успішно створено")
            except Exception as e:
                print(f"Неможливо створити директорію: {e}")
                return False
        else:
            print("Директорія існує")
        
        # Перевірка Graphviz
        try:
            результат = subprocess.run(['dot', '-V'], capture_output=True, text=True)
            print(f"Graphviz встановлено: {результат.stdout}")
        except FileNotFoundError:
            print("Помилка: Graphviz не знайдено! Встановіть його з https://graphviz.org/download/")
            return False
        
        # Перевірка прав доступу
        тест_файл = os.path.join(тест_директорія, 'тест.txt')
        try:
            with open(тест_файл, 'w') as f:
                f.write('тест')
            print("Тест запису: Успішно")
            
            with open(тест_файл, 'r') as f:
                вміст = f.read()
            print("Тест читання: Успішно")
            
            os.remove(тест_файл)
            print("Тест видалення: Успішно")
            
            return True
        except Exception as e:
            print(f"Помилка операції з файлом: {e}")
            return False
            
    except Exception as e:
        print(f"Помилка перевірки системи: {e}")
        return False

# Запуск перевірки
if перевірка_системи():
    print("\nВсі перевірки пройдено успішно ✅")
else:
    print("\nПеревірка системи не вдалася ❌")
```
